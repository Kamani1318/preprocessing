```python
import string,time
exclude = string.punctuation
import re
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
stopwords.words('english')
from nltk.stem.porter import PorterStemmer
```

```python
def remove_html_tags(text):
    return BeautifulSoup(text, 'html.parser').get_text()
```

```python
batch1_a_train = batch1_a_train.fillna('')
```

```python
def replace_punc(text):
    pattern = r'[^\w\s]'
    return re.sub(pattern, ',**,', text)
```

```python

def unmergetext(text):
    new_text = ""
    for char in text:
        if char.isupper() and new_text != "":
            new_text += " "
        new_text += char
    return new_text
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
def stringpreprocessing():
    new_text = ""
    t_my_list = []
    for i in range(1,BATCH_SIZE-9000):
        batch1_a_train.iloc[i,2] = replace_punc(remove_html_tags(batch1_a_train.iloc[i,2]))
        my_string = re.sub(r'[^\x00-\x7F]+', ',**,',batch1_a_train.iloc[i,2].lower())
        my_list = remove_stopwords(my_string).split(',**,')
        print(my_list)
        t_my_list.append(my_list)
    return t_my_list
```
